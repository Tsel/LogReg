```python
log10_x = np.log10(x)
```

```python
dM = np.zeros(total.size*4, dtype=int).reshape(28,4)
```

```python
dM[:7,0]=1
dM[7:14,1]=1
dM[14:21,2]=1
dM[21:28,3]=1
dM
```

```python
#
# discrete unit of group
m = len(set(group))
print m
```

```python
### Constant intercept model
with pm.Model() as co_in:
    beta_0 = pm.Normal('beta_0', 0., 1e-6)
    beta_  = pm.Normal('beta_', 0., 1e-6, shape=(m,))
    
    # expected parameter
    logit_p = (beta_0 + beta_[g]*dM[:,g] * T.log10(x))
    #
    p = T.exp(logit_p)/(1. + T.exp(logit_p))
    
    ### likelihood
    obs = pm.Binomial('obs', total, p, observed = pos)
```

```python
%%time
### Sampling
sample = 2000

with co_in:
    start = pm.find_MAP({
                         'beta_0' : -0.,
                         'beta_' : np.zeros(m)})
    step = pm.NUTS(scaling = start)
    
    trace = pm.psample(sample, step, start, progressbar=False, threads=4)
```

Finally we create an overlay for logit(p) given by the model for each of the categorial influential variables . The number of traces is known form above (threads=4).

```python
x = np.arange(0., 6.1, 0.1, dtype=float)
for tr in np.arange(4):
    t_ = trace.traces[tr]
    for p_ in np.arange(len(t_)):
        b_0 = (t_.point(p_)).get('beta_0')
        b_i = (t_.point(p_)).get('beta_')[0]
        z = b_0 + b_i * x
        prob = np.exp(z)/(1. + np.exp(z))
        plt.plot(x, prob, 'b-', alpha=0.01)
```

```python
import numpy as np, pymc as pm, theano.tensor as T, matplotlib.pyplot as plt
```

```python
#
# read data form file
fndata = "/Users/TOSS/Documents/Projects/PyMc3/LogitRegression/MDP_4_PyMc.csv"
data = np.genfromtxt(fndata, delimiter='\t', skip_header=1, usecols=np.arange(0,4))
data[:5,]
total = data[:,0]
pos   = data[:,1]
group = data[:,2]
x     = data[:,3]
```

```python
g = group.astype(int)
g
```

```python
import pandas as pd
```

```python
df = {'Total' : total,
      'Pos'   : pos,
      'Group' : group,
      'X (cont)': x}
```

```python
log10_x = np.log10(x)
```

```python
dM = np.zeros(total.size*4, dtype=int).reshape(28,4)
```

```python
dM[:7,0]=1
dM[7:14,1]=1
dM[14:21,2]=1
dM[21:28,3]=1
dM
```

```python
#
# discrete unit of group
m = len(set(group))
print m
```

```python
### Constant intercept model
with pm.Model() as co_in:
    beta_0 = pm.Normal('beta_0', 0., 1e-6)
    beta_  = pm.Normal('beta_', 0., 1e-6, shape=(m,))
    
    # expected parameter
    logit_p = (beta_0 + beta_[g]*dM[:,g] * log10_x)
    #
    p = T.exp(logit_p)/(1. + T.exp(logit_p))
    
    ### likelihood
    obs = pm.Binomial('obs', total, p, observed = pos)
```

```python
%%time
### Sampling
sample = 2000

with co_in:
    start = pm.find_MAP({
                         'beta_0' : 0.,
                         'beta_' : np.zeros(m)})
    step = pm.HamiltonianMC(scaling = start)
    
    trace = pm.psample(sample, step, start, progressbar=False, threads=4)
```

```python
burn = 1000
pm.summary(trace[burn:])
```

```python
from pymc import traceplot
traceplot(trace[burn:], vars = [beta_0, beta_]);
```

### 0 Generate some data for validation
Das binomiale logistische Regressionsmodell lautet:
$$
P(Y=1|X=x_i) = \frac{\exp(\beta_0 + x_i^T \mathbf{\beta})}{ 1 + \exp(\beta_0 + x_i^T \mathbf{\beta})} \qquad (1)
$$
wobei $\mathbf{\beta} = (\beta_1, \ldots, \beta_k)$.

Hier wird der einfache Fall $k=1$ untersucht, so dass

$$
m = \beta_0 + \beta_1\; x.
$$


Wahre Werte für $\beta_0$ und $\beta_1$ sind 10 und -0.5. Die kontinuierliche Einflussgröße $x$ ist aus dem Intervall (0, $\ldots$, $x_{max}, x_{max}=100$) .

Man erhält folgenden Zusammenhang aus (1).

```python
import numpy as np, pymc as pm, theano.tensor as T
```

```python
x_max = 100
x = np.linspace(0,x_max,200)
b0 = 10.
b1 = -0.5
P = np.exp(b0+b1*x)/(1 + np.exp(b0+b1*x))
```

```python
plt.plot(x,P)
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$P(Y=1|X=x)$", fontsize=18)
```

Nun werden $N=19$ binomial verteilte Zufallszahlen erzeugt, gemäß $y\sim Bin(n,p)$ mit $n=100$ und $p$ nach Gleichung (1). 

```python
n=100
N = 19
x_obs = np.linspace(0,x_max,19)
y_obs = np.zeros(len(x_obs))
for i in np.arange(len(x_obs)):
    m_ = b0 + b1*x_obs[i]
    p_ = (np.exp(m_)/(1.+np.exp(m_)))
    y_obs[i] = np.random.binomial(n,p_)
```

```python
print y_obs
```

```python
N_ = np.ones(N, dtype=int)*n
# model setup
###############################################
with pm.Model() as m:
    ## hyperpriors
    tau = pm.Gamma('tau', 1.e-3, 1.e-3)
    sigma = tau**-.5
    
    ### parameters
    # coefficients
    alpha1 = pm.Normal('alpha1', 0., 1e-6)
    alpha0 = pm.Normal('alpha0', 0., 1e-6)
    #
    b = pm.Normal('b', 0., tau, shape=(N,))
        
    # expected parameter
    logit_p = alpha0 + alpha1 * x_obs + b

    p = T.exp(logit_p) / (1 + T.exp(logit_p))
    
    ### likelihood
    obs = pm.Binomial('obs', N_, p, observed = y_obs)
```

```python
%%time
### Perform bayesian sampling

ssize=3000

with m:
    
    start = pm.find_MAP({'alpha0': 0.,
                     'alpha1': 0.,
                     'b' : np.zeros(N),
                     'tau' : 10.})
    step = pm.NUTS(scaling = start)
    #step = pm.HamiltonianMC(scaling = start)
    
    ptrace = pm.psample(ssize, step, start, threads=4) 
```

```python
burn = 1000
pm.summary(ptrace[burn:])
```

```python
from pymc import traceplot
traceplot(ptrace[burn:], vars = [alpha0, alpha1, tau]);
```
